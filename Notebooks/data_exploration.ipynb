```python
import polars as pl
```

```python
DATA_DIR  = "../data/ml-25m"          # folder containing ratings.csv, movies.csv, tags.csv
OUTPUT_DIR = "../data/eda_plots"
```

```python


ratings = pl.read_csv(f"{DATA_DIR}/ratings.csv",
                      schema_overrides={
                          "userId":    pl.Int32,
                          "movieId":   pl.Int32,
                          "rating":    pl.Float32,
                          "timestamp": pl.Int64
                      })

movies = pl.read_csv(f"{DATA_DIR}/movies.csv",
                     schema_overrides={
                         "movieId": pl.Int32,
                         "title":   pl.Utf8,
                         "genres":  pl.Utf8
                     })

tags = pl.read_csv(f"{DATA_DIR}/tags.csv",
                   schema_overrides={
                       "userId":    pl.Int32,
                       "movieId":   pl.Int32,
                       "tag":       pl.Utf8,
                       "timestamp": pl.Int64
                   })

# Add datetime column from Unix timestamp (seconds)
ratings = ratings.with_columns(
    pl.from_epoch("timestamp", time_unit="s").alias("datetime")
)
```
